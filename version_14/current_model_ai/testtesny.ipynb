```python

yolo_folder = '/home/student/Desktop/current_model_ai/obj_train_data'
images_folder = '/home/student/Desktop/current_model_ai/ImagesTotal'
output_folder = '/home/student/Desktop/current_model_ai'

# Output directories and clear existing labels
for subset in ['train', 'val', 'test']:
    images_path = os.path.join(output_folder, subset, 'images')
    labels_path = os.path.join(output_folder, subset, 'labels')

    os.makedirs(images_path, exist_ok=True)
    os.makedirs(labels_path, exist_ok=True)

    # Clear existing label files
    for file in os.listdir(labels_path):
        if file.endswith(".txt"):
            os.remove(os.path.join(labels_path, file))


# Collect annotation and their corresponding image
data_pairs = []

print("Searching for matching images and labels...")
for label_file in os.listdir(yolo_folder):
    if label_file.endswith(".txt"):
        label_path = os.path.join(yolo_folder, label_file)
        
        # Check for corresponding image 
        image_found = False
        for ext in ['.png', '.jpg', '.jpeg']:
            img_name = label_file.replace('.txt', ext)
            src_img = os.path.join(images_folder, img_name)
            if os.path.exists(src_img):
                data_pairs.append((src_img, label_path))
                print(f"Found: {img_name} with {label_file}")
                image_found = True
                break
        if not image_found:
            print(f"No matching image for label: {label_file}")

print(f"Total pairs found: {len(data_pairs)}")

# Shuffle and split data
random.shuffle(data_pairs)
num_images = len(data_pairs)
train_split = int(0.7 * num_images)
val_split = int(0.9 * num_images)

train_data = data_pairs[:train_split]
val_data = data_pairs[train_split:val_split]
test_data = data_pairs[val_split:]

# Save images and annotations in split
def save_data(data, subset):
    print(f"Saving {subset} data...")
    saved_count = 0
    for img_path, label_path in data:
        # Get filenames
        img_name = os.path.basename(img_path)
        label_name = os.path.basename(label_path)

        # Set destination paths
        dst_img = os.path.join(output_folder, subset, 'images', img_name)
        dst_label = os.path.join(output_folder, subset, 'labels', label_name)

        # Copy image and label
        shutil.copy(img_path, dst_img)
        shutil.copy(label_path, dst_label)
        saved_count += 1

    print(f"{saved_count} images and labels saved in {subset}.")

# Save to each subset
save_data(train_data, 'train')
save_data(val_data, 'val')
save_data(test_data, 'test')

print(f"Data split and saved successfully.")
```

```python
import os
import pandas as pd
import matplotlib.pyplot as plt

# Path to the folder containing tuning result CSVs
results_folder = '/home/student/Desktop/current_model_ai/folder_train/augment_tuning_scaled'

# Find all result CSVs, excluding the baseline file
result_files = [f for f in os.listdir(results_folder)
                if f.startswith('tuning_results_') and f.endswith('.csv')
                and 'no_augmentation' not in f]

# Store best values
best_values = []

for file in result_files:
    param_name = file.replace('tuning_results_', '').replace('.csv', '')
    path = os.path.join(results_folder, file)
    df = pd.read_csv(path)

    if not df.empty and 'mAP50-95' in df.columns:
        # Ensure the parameter column exists
        if param_name not in df.columns:
            print(f"Skipping {file}: missing column '{param_name}'")
            continue

        # Keep only rows where the parameter has a real value
        df_valid = df[df[param_name].notna()]
        df_valid_sorted = df_valid.sort_values(by=param_name)

        if df_valid_sorted.empty:
            print(f"No valid values for parameter '{param_name}'")
            continue

        # Find best row
        best_row = df_valid_sorted.loc[df_valid_sorted['mAP50-95'].idxmax()]
        best_values.append({
            'Parameter': param_name,
            'Best Value': best_row[param_name],
            'mAP50-95': best_row['mAP50-95'],
            'mAP50': best_row.get('mAP50'),
            'Precision': best_row.get('Precision'),
            'Recall': best_row.get('Recall')
        })

        # Plotting
        plt.figure()
        plt.plot(df_valid_sorted[param_name], df_valid_sorted['mAP50-95'], marker='o', label='mAP50-95')
        plt.title(f'mAP50-95 vs {param_name}')
        plt.xlabel(param_name)
        plt.ylabel('mAP50-95')
        plt.grid(True)
        plt.legend()
        plot_path = os.path.join(results_folder, f'plot_{param_name}.png')
        plt.savefig(plot_path)
        plt.close()

# Save best values summary
summary_df = pd.DataFrame(best_values)
summary_path = os.path.join(results_folder, 'best_augmentation_values.csv')
summary_df.to_csv(summary_path, index=False)

# Print results
print("Best augmentation values based on mAP50-95:")
print(summary_df)
```

```python
import os
import pandas as pd
import matplotlib.pyplot as plt

# Path to the folder containing tuning result CSVs
results_folder = '/home/student/Desktop/current_model_ai/folder_train/augment_tuning_scaled'

# Find all result CSVs (excluding no_augmentation file)
result_files = [f for f in os.listdir(results_folder)
                if f.startswith('tuning_results_') and f.endswith('.csv')
                and 'no_augmentation' not in f]

# Store best values
best_values = []

for file in result_files:
    param_name = file.replace('tuning_results_', '').replace('.csv', '')
    path = os.path.join(results_folder, file)
    df = pd.read_csv(path)

    if not df.empty and 'mAP50-95' in df.columns:
        if param_name not in df.columns:
            print(f"Skipping {file}: missing column '{param_name}'")
            continue

        # Keep only rows where the parameter value is not NaN and not 0.0
        df_valid = df[df[param_name].notna() & (df[param_name] != 0.0)]

        if df_valid.empty:
            print(f"No valid (non-zero) values for parameter '{param_name}'")
            continue

        df_valid_sorted = df_valid.sort_values(by=param_name)

        # Find best row
        best_row = df_valid_sorted.loc[df_valid_sorted['mAP50-95'].idxmax()]
        best_values.append({
            'Parameter': param_name,
            'Best Value': best_row[param_name],
            'mAP50-95': best_row['mAP50-95'],
            'mAP50': best_row.get('mAP50'),
            'Precision': best_row.get('Precision'),
            'Recall': best_row.get('Recall')
        })

        # Plotting without 0.0 values
        plt.figure()
        plt.plot(df_valid_sorted[param_name], df_valid_sorted['mAP50-95'], marker='o', label='mAP50-95')
        plt.title(f'mAP50-95 vs {param_name} (no 0.0)')
        plt.xlabel(param_name)
        plt.ylabel('mAP50-95')
        plt.grid(True)
        plt.legend()
        plot_path = os.path.join(results_folder, f'plot_{param_name}_no0.png')
        plt.savefig(plot_path)
        plt.close()

# Save summary of best values
summary_df = pd.DataFrame(best_values)
summary_path = os.path.join(results_folder, 'best_augmentation_values_no0.csv')
summary_df.to_csv(summary_path, index=False)

# Print summary
print("Best augmentation values (excluding 0.0) based on mAP50-95:")
print(summary_df)
```
